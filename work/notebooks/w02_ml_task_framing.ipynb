## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

I am choosing the **Ranking task** type. This is the optimal fit because our objective is not to classify content as "good" or "bad" as a static label, but to prioritize a queue of articles for an editor. Ranking provides a continuous score, translating directly into an actionable list that ensures the editor always tackles the highest-potential items first.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

My target is **future engagement growth**, defined as the delta in CTR (click-through rate) observed over a 30-day window following an editorial refresh. This is an observed outcome measured directly from our performance data, avoiding the biases inherent in rule-based labels (like is_declining). This allows the model to learn from real-world performance shifts rather than human-coded heuristics.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

My success metric is **Precision@K (top 5)**. Since an editor typically only has time to address a limited number of items, this metric measures whether the top 5 articles recommended by the model actually yield positive engagement growth. If an item from the "top 5" recommendations shows a lift after the editor's intervention, the model has successfully supported the decision.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule (e.g., "always update the article with the lowest CTR") fails because it ignores the context. A low CTR on an evergreen article is a signal for intervention, whereas a low CTR on a seasonal or news-based piece may be expected. The patterns are too tangled for static if-statements. ML allows us to weigh multiple signals—traffic source, scroll depth, and historical trends—dynamically and non-linearly to make a smarter prioritization decision than any static rule could provide.
